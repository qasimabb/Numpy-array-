```python
# Uniform Distribution is describe as probability where every event has equal Chances of its
# occurring for example if we take generation of random numbers so the probability of occurance 
# of any random number is almost same, it has 3 parameters
# a-lower bound its default value is 0
# b-upper bound its default value is 1
# size=The shape of an array which we need to return
```

```python
# Example Create a 2x3  uniform distribution sample 
import numpy as np
from numpy import random
x= random.uniform(size=(2,3))#print an array of 2 rows and every row have 3 elements
print(x) # print a 2D array with uniform distribution and random values
```

# Visualize Uniform Distribution

```python
import numpy as np
from numpy import random # random is module in numpy
import matplotlib.pyplot as plt # plotting graph 
import seaborn as sns # for dist plot
sns.distplot(random.uniform(size=1000),hist=False) # need to take uniform Dist, a size of 1000
                                                   # it will print 1D array in Uniform Dist
                                                      
plt.show()
```

# Logistic Distribution

```python
# it is use to describe the growth it is use in machine learning in logistic regression,neural 
# network etc , It has also 3 parameters
# loc-mean- where the peak present- it default value is 0
# scale-standard deviation ,flateness of distribution default value is 1
# size-The shape of the array we nwwd to return for this distribution
```

```python
# Draw a 2x3 samples from a logistic dist with mean at 1 and stddev 2.0
import numpy as np
from numpy import random
x=random.logistic(loc=1,scale=2.0,size=(2,3)) # 2 rows and each row have 3 values , will be 
print(x)
```

# Difference between  logistic and Normal Distribution

```python
# both dist are nearly identical, but logistic dist covered more area as compare to normal dist
# Actually it represent more possibility of occurrence of an event tahn the normal dist
# if we take higher value of scale(standard deviation) thenn normal and logistic dist are near
# identical apart from the peak(difference inthe peak) but over all same
```

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.normal(scale=2,size=1000),hist=False,label='normal') # for Normal dist
sns.distplot(random.logistic(size=1000),hist=False,label='logistic') # for logistic, by default
                                                                     # mean=0 , std = 1
plt.show()
# difference in peak
```

```python
# In binomial dist bi means we have 2 values
# for multinomial dist it is generalization of binomial dist it is come from binomial dist itself
# in binomial we have two outcomes but here we have multiple outcomes
# Blood type of a population A,B,O and so oon
# dice roll outcomes 1,2,3,4,5,6
# it has 3 parameters
#1 n-no of possible outcomes e.g (for dice 6 outcomes)
#2 pvals or probability values-list of probablities of outcomes every outcome will have certainly 
# a probablity(e.g[1/6,1/6,1/6,1/6,1/6,1/6]) for every outcome of dice roll
#3 size-shape of returned array
```

```python
#Draw out a sample of dice roll
import numpy as np
from numpy import random
x = random.multinomial(n=6 ,pvals=[1/6,1/6,1/6,1/6,1/6,1/6]) # n=6 six trial , pval =
                                                            # probab of each trial
print(x)
```

```python
# Note
# multinomial sample will not produce a single value actually they will produce one value for
# each pval(every outcome assosiated with  the probability)
#(it is also generalize form of binomial dist)if we talk about visual representation of multinom
# dist it will be similar to multiplebinomial dist if it is compare to normal distribution
```

# Visualization of Multinomial Distribution

```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
sns.distplot(random.multinomial(n=6,pvals=[1/6,1/6,1/6,1/6,1/6,1/6],size=10),hist=False, label= 'multinomial')
# no of trials=6 and every respect value we have probab , sample size = 10
plt.show()
```
